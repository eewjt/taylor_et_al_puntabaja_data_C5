```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format='retina'
```

```python
df=pd.read_csv ('data/pb_bedforms.csv')
```

```python
sns.jointplot(data = df, x=df.wavelength, y=df.amplitude, hue=df.bedform_name)
plt.xlim(-50,650)
plt. ylim(-10,75)
plt.savefig('figs/bedforms_names.tif', dpi=300)
```

```python
sns.jointplot(data = df, x=df.wavelength, y=df.amplitude, hue=df.bedform_type)
plt.xlim(-50,650)
plt. ylim(-10,75)

plt.savefig('figs/bedforms_type.tif', dpi=300)
```
