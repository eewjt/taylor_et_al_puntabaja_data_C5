```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format='retina'
sns.set_style('ticks')
```

```python
df=pd.read_csv ('data/pb_bedforms.csv')
```

```python
# Filter data

type_a = df[(df['bedform_type'] == 'a') | (df['bedform_type'] == 's')]
```

```python
sns.scatterplot(data = df, x='wavelength', y='amplitude', hue='bedform_name')


plt.savefig('figs/bedforms_names.eps', dpi=300)
```
